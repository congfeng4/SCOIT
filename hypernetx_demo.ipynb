```python
# First, install the hypernetx library if you haven't already
# You can install it using pip:
# pip install hypernetx

import hypernetx as hnx
import matplotlib.pyplot as plt

# Define the hyperedges
# Each hyperedge is a tuple of nodes it connects
# For example, let's create a hypergraph with nodes A, B, C, D
# and hyperedges connecting (A, B, C), (B, D), and (C, D)
edges = [
    ('A', 'B', 'C'),  # Hyperedge 1: connects A, B, C
    ('B', 'D'),       # Hyperedge 2: connects B, D
    ('C', 'D')        # Hyperedge 3: connects C, D
]

# Create the hypergraph from the edges
H = hnx.Hypergraph(edges)

# Print information about the hypergraph
print("Hypergraph H:")
print(f"Nodes: {list(H.nodes)}")
print(f"Hyperedges: {list(H.edges)}")

# Visualize the hypergraph
# Note: Visualization might not work in all environments
# but should work in Jupyter notebooks or environments with matplotlib support
try:
    hnx.draw(H)
    plt.title("Hypergraph Visualization")
    plt.show()
except Exception as e:
    print(f"Visualization failed: {e}")
    print("The hypergraph was created successfully but visualization may not be supported in this environment.")
```

```python
from ipywidgets import interact

@interact(level=(0, 5, 1))
def show_level(level):
    sub = hnx.algorithms.simplification.simplify(H, level=level)  # 官方简化 API
    hnx.draw(sub, layout='matrix', node_size=1, edge_color='tab:blue')
    plt.show()
```
